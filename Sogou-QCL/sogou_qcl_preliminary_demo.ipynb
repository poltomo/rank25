# online tokenizer demo

```python
import math

dat1 = next(stream_sogou_qcl3("./main_v2/SogouQCL.0.v2.bz2"))
_, docs, _ = dat1
_, doc1 = docs[0]
chinese_text = doc1['content']
print(chinese_text)

tokenizer = OnlineJiebaWordLevelTokenizer()
tokenizer.train_on_chunk(chinese_text, min_frequency=1, top_k_new_tokens=100, length_preference_factor=1)
token_ids = tokenizer.tokenize(chinese_text)
print(token_ids)
print(tokenizer.decode(token_ids=token_ids))
```

# RocksDB demo

```python
from torch_geometric.data import RocksDatabase

import shutil
import gc

db_path = "./docs/"
db = RocksDatabase(db_path)

_, docs, _ = next(stream_sogou_qcl3("./main_v2/SogouQCL.0.v2.bz2"))

my_doc_id = docs[0][0]

try:
    print(db.get(my_doc_id))
except:
    print(f"no such key: {my_doc_id}")

for (doc_id, doc) in docs:
    db.insert(doc_id, doc)

try:
    print(db.get(my_doc_id))
except:
    print(f"no such key: {my_doc_id}")

db.close()
shutil.rmtree(db_path,)
del db
gc.collect()
```
